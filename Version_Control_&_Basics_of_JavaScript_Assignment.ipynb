Question 1 : Explain the purpose of version control systems like Git. Why are they
essential in modern software development?

Version control systems like Git are essential because they track every change in a project, allow multiple developers to collaborate seamlessly, and ensure that code can always be restored to a previous working state.

 Purpose of Version Control Systems :


Track changes: Every modification to the code is recorded, including who made it, when, and why. This creates a transparent history of the project.

Collaboration: Multiple developers can work on different features or bug fixes simultaneously without overwriting each other’s work.

Experimentation: Teams can create branches to test new ideas safely, merging only when stable.

 Why Git is Essential in Modern Development :


Distributed system: Every developer has a full copy of the project and its history, enabling offline work and reducing reliance on a central server.

Speed & efficiency: Git operations are fast, lightweight, and optimized for large projects.

Supports DevOps: Git integrates with CI/CD pipelines, enabling automated testing, deployment, and continuous delivery.

Question 2: List the typical steps you would follow when working with Git to make
changes to a project and share them on GitHub.


Clone the repository:

Download a local copy of the project from GitHub using `git clone <repo_url>`.

Create a new branch:

Open a separate branch for your changes with `git checkout -b branch_name` to keep work isolated.

Make and stage changes:

Edit files as needed, then add them to the staging area using `git add <file>`.

Commit your changes:  

Save your work with a descriptive message using `git commit -m "message"`.

Push branch to GitHub:

Upload your branch to the remote repository with `git push origin branch_name`.

Open a pull request :

On GitHub, create a pull request to merge your branch into the main project.

Key Notes :

Branching keeps your work separate until it’s ready to merge.

Commit messages should be clear and descriptive to help track history.

Question 7: You are given the following array:
const colors = ["red", "green", "blue", "yellow", "purple"];
1. Access and print the first element of the array.
2. Access and print the last element of the array.
3. Access and print the third element of the array.


```python
const colors = ["red", "green", "blue", "yellow", "purple"];

// 1. Access and print the first element
console.log(colors[0]);   // Output: red

// 2. Access and print the last element
console.log(colors[colors.length - 1]);   // Output: purple

// 3. Access and print the third element
console.log(colors[2]);   // Output: blue
```

Key Notes :

Array indexing starts at 0, so the first element is colors[0].

Last element access uses colors.length - 1.



Question 8: Given x = 10 and y = 5, write expressions to:
1. Add x and y
2. Check if x is greater than y
3. Check if x is equal to y

```python
let x = 10;
let y = 5;

// 1. Add x and y
console.log(x + y);   // Output: 15

// 2. Check if x is greater than y
console.log(x > y);   // Output: true

// 3. Check if x is equal to y
console.log(x === y); // Output: false
```

Key Notes :

Addition operator (+) combines values.

Comparison operator > checks if one value is larger than another.



Question 9: Identify the data types of the following values:
1. "Hello"
2. 42
3. true
4. undefined

```python
// 1. "Hello"
typeof "Hello";   // Output: "string"

// 2. 42
typeof 42;        // Output: "number"

// 3. true
typeof true;      // Output: "boolean"

// 4. undefined
typeof undefined; // Output: "undefined"
```

Key Notes :

"Hello" → string

42 → number

true → boolean
